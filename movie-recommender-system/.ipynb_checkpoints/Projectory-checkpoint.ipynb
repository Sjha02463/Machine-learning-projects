```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies.head(2)
```

```python
credits.head()
```

```python
movies.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
#genre
#id
#keywords
#title
#overview
#cast
#crew
#we will avoid the numerical datas

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append (i['name'])
    return l
```

```python
movies.dropna(inplace = True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#In the above the format of keywords and genres have been changed now we are choosing here the top 3 casts
```

```python
movies['cast'][0]
```

```python
def convert3(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
           l.append(i['name'])
           counter+=1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#After choosing top three casts we have to choose the crew member in which the job is director
```

```python
def fetch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append (i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#in the crew we have fetched only the directors name and now we have to convert the overview from string to lists
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Now we have to remove the spaces between two words as to make them a single tag only DATA PREPROCESSING
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
#We have already created  the tags so now we can remove the overview, genres,keywords,cast,crew
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
#In this we have converted the letters into the lower case
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
#Vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray() #It will convert the sparse matrix to numpy array
```

```python
#Here many similar words are present there so we can remove them by the process of steaming
#For Example
#['loved','loving,'love'] After applying steaming we will get the words ['love', 'love', 'love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
#Here porterStemmer is a class and now we have to make a class
ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))#steming for each word
    return " ".join(y)#we are again converting the lists into String 
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#4806 movies and thus we cn say thet there are 4806 vectors and in each vectors there are 5000 numbers
# We have 5000 dimensional spaces and we have to calculate the distances of each movie with another movie and the one with the closest distance will be recommended
#More will be the distance then less will be the similarities...and we will not calculate the euclidean distance rather we will calculate the cosine distances as in the vector we use the cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    # Ensure the movie exists in the dataset
    if movie not in new_df['title'].values:
        print(f"Movie '{movie}' not found in dataset.")
        return []
    
    # Fetch the index of the movie correctly
    movie_index_list = new_df[new_df['title'] == movie].index

    if len(movie_index_list) == 0:
        print(f"Movie '{movie}' not found in dataset.")
        return []

    movie_index = movie_index_list[0]  # Get the first matching index

    distances = similarity[movie_index]

    # Get top 5 similar movies (excluding itself)
    recommended_indices = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    # Fetch movie titles
    recommended_movies = [new_df.iloc[i[0]]['title'] for i in recommended_indices]
    
    for movie_title in recommended_movies:
        print(movie_title)
```

```python
# Test the function
recommend("Avatar")
```
